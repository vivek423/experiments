```python
import pandas as pd
import numpy as np

import scipy.stats as stats
from scipy.stats import shapiro,normaltest

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
results = pd.read_csv('landing_page.csv')
```

```python
results.head()
```

## Visually observe boxcox to guage if normal distribution holds in case of small sample

```python
# generate a boxplot to see the data distribution by treatments. Using boxplot, we can easily detect the differences 
# between different treatments
results.boxplot(column=['Rake'], by=['adkey'],figsize=(20,8))
plt.show()
```

```python
results['Rake'].hist(by=results['adkey'],figsize=(20,10),bins=50)
```

## If not normal, in most cases taking the log works. Especially in case of continuous monetory metrics - Rake, Costs, Winnings, etc

```python
# generate a boxplot to see the data distribution by treatments. Using boxplot, we can easily detect the differences 
# between different treatments
results['ln_rake'] = np.log(results['Rake'])
results.boxplot(column=['ln_rake'], by=['adkey'],figsize=(20,8))
plt.show()
```

```python
results['ln_rake'].hist(by=results['adkey'],figsize=(20,10),bins=50)
```

## To double check, run normality test

```python
def normality_test(metric,variant):
    data = results[results['adkey']==variant]
    data = np.array(data[metric].values)
    stat, p = normaltest(data)
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    # interpret
    alpha = 0.05
    if p > alpha:
        print(variant,':','Sample looks Gaussian (fail to reject H0)','\n')
    else:
        print(variant,':','Sample does not look Gaussian (reject H0)','\n')
```

```python
normality_test('ln_rake','Mobile_HPheaderJulySet1_button')
normality_test('ln_rake','Mobile_HPheaderJulySet2_button')
normality_test('ln_rake','Mobile_HPheaderJulySet4_button')
normality_test('ln_rake','Mobile_HPheaderSrini_button')
```

## Finally run t-test and conclude

```python
def run_ttest(metric,variant1,variant2):
    sample1 = results[results['adkey']==variant1]
    sample1 = np.array(sample1[metric].values)

    sample2 = results[results['adkey']==variant2]
    sample2 = np.array(sample2[metric].values)

    t_stat, p_val = stats.ttest_ind(sample1, sample2, equal_var=False)
    
    if p_val > 0.05:
        significance = 'Sample means are same'
    else:
        significance = 'Sample means are different'

    print('Variant1:',variant1,'Variant2:',variant2,significance,'p_value = ',p_val,'\n')
```

```python
run_ttest('ln_rake','Mobile_HPheaderJulySet1_button','Mobile_HPheaderSrini_button')
run_ttest('ln_rake','Mobile_HPheaderJulySet2_button','Mobile_HPheaderSrini_button')
run_ttest('ln_rake','Mobile_HPheaderJulySet4_button','Mobile_HPheaderSrini_button')
```
